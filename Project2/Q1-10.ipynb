```r
#Q5,6,7
options(repr.plot.width=5,repr.plot.height=5)

# create a sub graph with 1-neighborhood of node 1
personal_network = induced.subgraph(g, c(1, neighbors(g,1)))
vertexvector = rep(3.5,vcount(personal_network))#this gives size of 3.5 to all neighbor nodes
vertexvector[1]=6 # this specifies teh size of the core node as 6
vertexcolor = rep("green",vcount(personal_network)) #this gives color to all neighbor nodes
vertexcolor[1] ="red" #color of teh core node
plot.igraph(personal_network,vertex.size=vertexvector,vertex.label =NA,vertex.color=vertexcolor)

# number of nodes
node_count = vcount(personal_network)
cat("Number of Nodes ", node_count)

# number of edges
edge_count = ecount(personal_network)
cat("\nNumber of edges ", edge_count)

dia= diameter(personal_network)
cat("\nDiameter of personal network ", dia)
```

```r
# Finding core nodes Q8
core_nodes <- numeric()
for (each_vertex in V(g)) {
  if (length(neighbors(g, each_vertex)) >= 200)
    core_nodes <- c(core_nodes, each_vertex)
}
```

```r
#Q8b Average degree of core nodes
cdeg<-degree(g,core_nodes)
```

```r
#Q9  For each of these core nodes, generate the community
node_list=c(1,108,349,484,1087)
setwd("./plots")
for(node in node_list){

    cat("****************For node : ",node,"************************************\n")
    #generate personalized network for this core node
    neighborhood = neighborhood(g, order=1, nodes=node)
    personal_network = induced.subgraph(g, vids=unlist(neighborhood), impl="auto")
    personal_network$label = sort(unlist(neighborhood))
    #generate fastgreedy community
    fg <- fastgreedy.community(personal_network)
    #cmsize <- sizes(fg)
    #print(cmsize)
    
    #modularity is 
    cat("Fastgreedy Modularity is ",modularity(fg),"\n")
    jpeg(paste(node,'_fgplot.jpg',sep = ""))
    # jpeg(paste(node,'fgplot.jpg'))
    plot(fg, personal_network, edge.color = "light grey", vertex.label=NA, vertex.size=3, asp=12/16)
 dev.off()

    print("_______________________________________________________________")
    #generate edge betweenness community
    
    eg <- edge.betweenness.community(personal_network)
    #cmsize <- sizes(eg)
    #print(cmsize)
    #modularity is 
    cat("Edge betweenness  Modularity is ",modularity(eg),"\n")
jpeg(paste(node,'_egplot.jpg',sep = ""))    
#jpeg('egplot.jpg')
    plot(eg, personal_network, edge.color = "light grey", vertex.label=NA, vertex.size=3, asp=12/16)
 dev.off()

    print("_______________________________________________________________")
    #generate Infomap community
    
    ic=cluster_infomap(personal_network)
jpeg(paste(node,'_icplot.jpg',sep = ""))    
#jpeg('icplot.jpg')
    plot(ic, personal_network, edge.color = "light grey", vertex.label=NA, vertex.size=3, asp=12/16)
 dev.off()

    cat("Infomap  Modularity is ",modularity(ic),"\n")

    
}
```
