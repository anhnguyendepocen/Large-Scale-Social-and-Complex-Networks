```r
triFile = read.table("project5/tri.txt",sep="\t",quote="")

elist=c()
for(i in 1:nrow(triFile)) {
    row <- triFile[i,]
    v1=row$"V1"
    v2=row$"V2"
    v1=which(vertex_attr(g.giant)$name==v1)
    v2=which(vertex_attr(g.giant)$name==v2)
    
    if(!are.connected(g.giant,v1,v2)){
        e<-(shortest.paths(g.giant,v1,v2))
        g.giant=add_edges(g.giant,c(v1,v2),weight=e)
        }
    ei<-get.edge.ids(g.giant,c(v1,v2), error = FALSE, multi = FALSE)
    elist<-c(elist,ei)
}
```

```r
find_rate <- function(v1,v2) {
    
x1=jsonFile[which(jsonFile$"V1"==v1),]['V3']
y1=jsonFile[which(jsonFile$"V1"==v1),]['V4']
x2=jsonFile[which(jsonFile$"V1"==v2),]['V3']
y2=jsonFile[which(jsonFile$"V1"==v2),]['V4']
d=sqrt((y2-y1)^2+(x2-x1)^2)
d=d*69   
ei<-get.edge.ids(g.giant,c(v1,v2), error = FALSE, multi = FALSE)
t<-E(g.giant)$weight[ei]
t1<-(shortest.paths(g1,v1,v2))

cs=d/t
    
fr=3600/((0.003/cs)+2)
return(fr)
}
```

```r
find_rateg2 <- function(v1,v2) {
    
x1=jsonFile[which(jsonFile$"V1"==v1),]['V3']
y1=jsonFile[which(jsonFile$"V1"==v1),]['V4']
x2=jsonFile[which(jsonFile$"V1"==v2),]['V3']
y2=jsonFile[which(jsonFile$"V1"==v2),]['V4']
d=sqrt((y2-y1)^2+(x2-x1)^2)
d=d*69   
ei<-get.edge.ids(g2,c(v1,v2), error = FALSE, multi = FALSE)
t<-E(g2)$weight[ei]
t1<-(shortest.paths(g2,v1,v2))

cs=d/t
    
fr=3600/((0.003/cs)+2)
return(fr)
}
```
